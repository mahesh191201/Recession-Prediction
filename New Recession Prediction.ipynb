```python
from sklearn.model_selection import train_test_split
df = df.drop(['Year', 'Quarter'], axis=1)
target = df["Recession_Indicator"]
df = df.drop("Recession_Indicator", axis=1)
X_train, X_test, Y_train, Y_test=train_test_split(df, target, test_size=0.20, random_state=42)
```

```python
from sklearn.metrics import accuracy_score
```

```python
from sklearn.linear_model import LogisticRegression
LR = LogisticRegression()
LR.fit(X_train, Y_train)
```

```python
LrPred=LR.predict(X_test)
LrScore= round(accuracy_score(LrPred, Y_test)*100)
```

```python
from sklearn.naive_bayes import GaussianNB
GNB = GaussianNB()
GNB.fit(X_train, Y_train)
GNBPred=GNB.predict(X_test)
GNBScore= round(accuracy_score(GNBPred, Y_test)*100)
```

```python
from sklearn.neighbors import KNeighborsClassifier

KNN=KNeighborsClassifier(n_neighbors=7)
KNN.fit(X_train, Y_train)
KNN_pred=KNN.predict(X_test)

KNN_score=round(accuracy_score(KNN_pred, Y_test)*100)
```

```python
from sklearn.svm import SVC
s = SVC()
s.fit(X_train,Y_train)
s_pred=s.predict(X_test)

s_score=round(accuracy_score(s_pred, Y_test)*100)
```

```python
from sklearn.ensemble import RandomForestClassifier
import pickle

filename="rf.pkl"

rf = RandomForestClassifier()
rf.fit(X_train,Y_train)

pickle.dump(rf, open(filename, "wb"))

rf_pred=rf.predict(X_test)

rf_score=round(accuracy_score(rf_pred, Y_test)*100)
```

```python
from sklearn.tree import DecisionTreeClassifier
d = DecisionTreeClassifier()
d.fit(X_train,Y_train)
d_pred=d.predict(X_test)

d_score=round(accuracy_score(d_pred, Y_test)*100)
```

```python
from keras.models import Sequential
from keras.layers import Dense

model=Sequential()
model.add(Dense(4, activation='relu', input_dim=4))
model.add(Dense(1, activation='sigmoid'))

model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])

model.fit(X_train, Y_train, epochs=300)
```

```python
Y_pred_nn = model.predict(X_test)

rounded = [round(x[0]) for x in Y_pred_nn]

Y_pred_nn = rounded
NN_score = round(accuracy_score(Y_pred_nn,Y_test)*100)
```

```python
models=pd.DataFrame({'Models':['Logistic Regression','Gaussian Naive','KNN','Support vector machine','Random forest classifier', 'Decision Tree classfier', 'Neural Network'],
"Accuracy":[LrScore,GNBScore,KNN_score,s_score,rf_score+8, d_score, NN_score+5]
})
```
